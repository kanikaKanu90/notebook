```
def fill_missing(df):
    for row in range(df.shape[0]):
        for col in range(df.shape[1]):
            if np.isnan(df[row, col]):
                # print("nan for row ", row, " col ", col)
                # print("value to be assigned: ", df[row-1, col], " to ", df[row, col])
                df[row, col] = df[row-1, col]        

fill_missing(df.values)

np.isnan(df).sum()
```

```
df.to_csv('/content/sample_data/exported.csv', index = True)
```

```
daily_changes_df = pd.DataFrame()

def calculate_daily_changes(data_toBe_treated):
  for row in range(data_toBe_treated.shape[0]):
    for col in range(data_toBe_treated.shape[1]):
      daily_changes_df.loc[row, col] = data_toBe_treated[row, col] - data_toBe_treated[row-1, col]


calculate_daily_changes(df.values)

daily_changes_df.fillna(0)
```

```
daily_changes_df.to_csv('/content/sample_data/changes_daily.csv', index = True)
```

```
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource
from bokeh.models.tools import HoverTool

from bokeh.palettes import Spectral5
from bokeh.transform import factor_cmap

output_file('cases_by_date.html')

grouped = df.groupby('Date')['AG', 'AI', 'AR', 'BE', 'BL', 'BS', 'FR', 'GE', 'GL', 'GR',
        'VS_pc', 'VS_diff_pc', 'ZG_diff', 'ZG_pc', 'ZG_diff_pc', 'ZH_diff',
        'ZH_pc', 'ZH_diff_pc', 'CH_diff', 'CH_pc'].sum()

print(grouped)        

source = ColumnDataSource(grouped)
cantons = source.data['Date'].tolist()
p = figure(x_range=cantons) #figure(x_axis_type="datetime")

color_map = factor_cmap(field_name='Date',
                    palette=Spectral5, factors=cantons)

p.vbar(x='Date', top='AG', source=source, width=0.70, color=color_map)

p.title.text ='Cases as per Canton'
p.xaxis.axis_label = 'Canton'
p.yaxis.axis_label = 'Date'

hover = HoverTool()
hover.tooltips = [
    ("Totals", "@AG / @AI / @AR")]

hover.mode = 'vline'

p.add_tools(hover)

show(p)
```
