```
import pandas as pd
import numpy as np

df = pd.read_csv("/content/sample_data/covid19_cases_switzerland_openzh-phase2.csv")
```

```
df.set_index(['Date'],inplace=True) # set index
```

```
df.shape
```

```
df.drop(columns=['CH_diff_pc'],inplace=True, axis=1)
```

```
df_temp = pd.DataFrame(data= np.zeros(shape=(1,107), dtype= 'float64'), columns=list(df.columns))
df = df_temp.append(df, ignore_index= True)
df.dtypes
```

```
df.shape
df.head()
```

```
np.isnan(df).sum()
```

```
df.shape
```

```
def fill_missing(df):
    for row in range(df.shape[0]):
        for col in range(df.shape[1]):
            if np.isnan(df[row, col]):
                print("nan for row ", row, " col ", col)
                print("value to be assigned: ", df[row-1, col], " to ", df[row, col])
                df[row, col] = df[row-1, col]        

fill_missing(df.values)

np.isnan(df).sum()
```

```
df.shape
```

```
df.to_csv('/content/sample_data/exported.csv', index = True)
```

```
daily_changes_df = pd.DataFrame(data= np.zeros(shape=(149,107), dtype= 'float64'), columns=list(df.columns))

def calculate_daily_changes(df):
    for row in range(df.shape[0]):
        if row!=0:
          for col in range(df.shape[1]):
                daily_changes_df[row, col] = df[row, col] - df[row-1, col]


calculate_daily_changes(df.values)
daily_changes_df
```

```
daily_changes_df.to_csv('/content/sample_data/changes_daily.csv', index = True)
```
